```python
from scipy.fftpack import fft  
import matplotlib.pyplot as plt 
import numpy as np  

x1 = np.random.rand(20) #Se define variable X1 que se mantiene en ambas gráficas con el método random, obteniendo 20 datos aleatorios
y1 = np.random.rand(20)  #Se define variable Y1 para la primera gráfica con el método random, obteniendo 20 datos aleatorios
y2 = fft(x1) #Se define variable Y2 para la segunda gráfica aplicando la transformada de Fourier a los datos de X1

fig, (ax1,ax2) = plt.subplots(1,2, figsize = (12,8)) #Se crean ambas gráficas, se define el orden y el tamaño de presentación.

ax1.plot(x1,y1, 'r', label = 'Datos',color='blue', linewidth = 3) #Se define la primera gráfica con los datos de X1 y Y1 (x,y), con formato de línea, color azul, nombre de la leyenda: 'Datos' y el grosor de la línea '3'
ax2.plot(x1,y2, 'o',label = "D. transformado",color='red', linewidth = 6) #Se define la segunda gráfica con los datos de X1 y Y2 (x,y), con formato de puntos, color rojo, nombre de la leyenda: 'D. transformado' y el grosor de puntos '6'

ax1.set_xlabel("Tiempo") # Se asigna título al eje x de la primera gráfica
ax1.set_ylabel("Amplitud") # Se asigna título al eje y de la primera gráfica
ax1.set_title("Señal sin transformar") # Se asigna título de la primera gráfica

ax2.set_xlabel("Frecuencia") # Se asigna título al eje x de la segunda gráfica
ax2.set_ylabel("Amplitud") # Se asigna título al eje y de la segunda gráfica
ax2.set_title("Señal transformada") # Se asigna título de la segunda gráfica

ax1.legend() # Se muestra leyenda de primera gráfica
ax2.legend() # Se muestra leyenda de segunda gráfica
```
