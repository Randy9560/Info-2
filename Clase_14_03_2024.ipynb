En esta clase se van a ver las librerias de matplotlib y seaborn, estos sirven para crear graficos


Para descarcargar Matplotlib escribimos en la terminal "py -m pip install matplotlib"

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = np.linspace(0, 2*np.pi,200)
x2 = np.linspace(0, 3*np.pi,200)


y1 = np.sin(x1)
y2 = np.cos(x2)

fig, (ax1,ax2) = plt.subplots(2,1, figsize = (10,10))
ax1.plot(x1,y1, 'v', label = "Grafica sen(x)",color='red', linewidth = 3)#, linestyle = "--")
ax2.plot(x2,y2, 'o', label = "Grafica cos(x)",color='black', linewidth = 4)#, linestyle = ":")
ax1.set_xlabel("Eje x")
ax1.set_ylabel("Eje y")
ax1.set_title("Grafica para la funcion sen(x)")
ax2.set_xlabel("Eje x")
ax2.set_ylabel("Eje y")
ax2.set_title("Grafica para la funcion cos(x)")
ax1.legend()
ax2.legend()

plt.show()
```

Par crear una figura en la que se especifiquen claramente los Axes se utilizan para crear la grafica

```python
fig, ax = plt.subplots() #Metodo para hacer una grafica
ax.plot([1,2,3,4],[1,4,2,3]) #La priemra entrada es x la segunda es y
plt.show()
```

Partes de una figura.

Estos son los compoenentes de una figura de Matplotlib:

-caudricula;
-titulo;
-ancho de linea;
-graficar formas;
-se puede tener una leyende; 
-min y max de valores de ejes que voy a mostrar;

------------------------------------------------------------------------------------

Los ejes tienen se pueden modificar por metodos

set_title() --> titulo de la grafica;
set_xlabel() --> titulo del eje x;
set_ylabel() --> titulo de eje y.
