1. SISTEMA DE GESTIÓN DE SIGNOS VITALES

Se propone diseñar un sistema para la gestión de signos vitales de pacientes en un servicio de hospitalización. El sistema debe gestionar médicos y pacientes. Los pacientes tienen información de nombre, cédula, temperatura, frecuencia cardiaca, presión. Los médicos tienen información de nombre, cédula, registromédico y la lista de los pacientes que tiene a cargo. El médico puede verificar el estado de un paciente o ingresar un nuevo paciente. Para verificar el estado de unpaciente se ingresa su número de cédula y el sistema deberá retornar una cadena con el siguiente formato:
Nombre; Temperatura; Frecuencia cardiaca; Presión, en el orden indicado y usando los ‘;’. En caso de que el
número de identificación no exista se debe retornar la cadena: Paciente no existe. Para ingresar un paciente se solicita su identificación, en caso de que exista se retorna la cadena: Paciente
existe. En el otro caso se ingresa los datos restantes, y una vez creado se agrega a la lista 

```python
class Persona:
    def __init__(self, nombre, cedula):
        self.__nombre = self.validar_nombre(nombre)
        self.__cedula = self.validar_cedula(cedula)

    def validar_nombre(self, nombre):
        if not isinstance(nombre, str):
            raise ValueError("El nombre tiene que ser un string")
        return nombre

    def validar_cedula(self, cedula):
        if not isinstance(cedula, int) or cedula < 0:
            raise ValueError("La cedula tiene que estar en numeros entero")
        return cedula

    @property
    def nombre(self):
        return self.__nombre
    
    @property
    def cedula(self):
        return self.__cedula

    @nombre.setter
    def nombre(self, nuevoNombre):
        self.__nombre = self.validar_nombre(nuevoNombre)

    @cedula.setter
    def cedula(self, nuevaCedula):
        self.__cedula = self.validar_cedula(nuevaCedula)
    
    def __str__(self):
        return f""" El nombre es: {self.nombre}, 
                    El N° de cedula es: {self.cedula}"""

class Paciente(Persona):
    def __init__(self, nombre, cedula, temperatura, frecuenciaCardiaca, presion):
        super().__init__(nombre, cedula)
        self.__temperatura = self.validar_temperatura(temperatura)
        self.__frecuenciaCardiaca = self.validar_frecuenciaCardiaca(frecuenciaCardiaca)
        self.__presion = self.validar_presion(presion)
        Medico.adicionarCedula(cedula)
        Medico.adicionarPaciente(self)

    # def __eq__(self, other):
    #     if isinstance(other, Paciente):
    #         return self.cedula == other.cedula
    #     return False

    def validar_temperatura(self, temperatura):
        if not isinstance(temperatura, (float,int)):
            raise ValueError("La temperatura debe ser un numero")
        return temperatura

    def validar_frecuenciaCardiaca(self, fcardiaca):
        if not isinstance(fcardiaca, (float,int)) or fcardiaca < 0:
            raise ValueError("La frecuencia Cardiaca debe ser un numero")
        return fcardiaca

    def validar_presion(self, presion):
        if not isinstance(presion, (float,int)) or presion < 0:
            raise ValueError("La presion debe ser un numero")
        return presion

    @property
    def temperatura(self):
        return self.__temperatura
    @temperatura.setter
    def temperatura(self, nuevatemperatura):
        self.__temperatura = self.validar_temperatura(nuevatemperatura)

    @property
    def frecuenciaCardiaca(self):
        return self.__frecuenciaCardiaca
    @frecuenciaCardiaca.setter
    def frecuenciaCardiaca(self, nuevafcardaica):
        self.__frecuenciaCardiaca = self.validar_frecuenciaCardiaca(nuevafcardaica)

    @property
    def presion(self):
        return self.__presion
    @presion.setter
    def presion(self, nuevaPresion):
        self.__presion = self.validar_presion(nuevaPresion)

    def __str__(self):
        return f"""{super().__str__()}, 
                   La temperatura es: {self.temperatura},
                   La frecuencia Cardiaca es: {self.frecuenciaCardiaca},
                   La presion es: {self.presion}
                   """    

class Medico(Persona):
    listaPacientes = []
    lista_cedula_pacientes = []
    def __init__(self, nombre, cedula, registroMedico):
        super().__init__(nombre, cedula)
        self.registroMedico = registroMedico

    @classmethod
    def adicionarPaciente(cls, paciente):
        print(paciente in cls.listaPacientes)
        if paciente in cls.lista_cedula_pacientes:
            print("El paciente ya ha sido ingresado")
        else:
            cls.listaPacientes.append(paciente)

    @classmethod
    def adicionarCedula(cls, cedula):
        if cedula in cls.lista_cedula_pacientes:
            print("La cedula del paciente ya ha sido ingresada")
        else:
            cls.lista_cedula_pacientes.append(cedula)

    def verificarPaciente(self, cedula):
        if cedula in Medico.lista_cedula_pacientes:
            for i in Medico.listaPacientes:
                if cedula == i.cedula:
                    print(f'Nombre: {i.nombre}; \nTemperatura: {i.temperatura}; \nFrecuencia cardiaca: {i.frecuenciaCardiaca}; \nPresión: {i.presion}')
        
        elif cedula not in Medico.lista_cedula_pacientes:
            print("Paciente no existe.")

    def agregarPaciente(self, cedula):
        if cedula in Medico.lista_cedula_pacientes:
            print("Paciente existe")

        else:
            nombre = input("Ingrese en nombre del paciente: ")
            temperatura = float(input("Ingrese la temperatura del paciente: "))
            fCardiaca = float(input("Ingrese la frecuencia cardiaca del paciente: "))
            presion = float(input("Ingrese la presion del paciente: "))
            Paciente(nombre, cedula, temperatura, fCardiaca, presion)

         

try:
    paceinte1 =  Paciente("jhon",123,20.0,52.0,4.5)
    # print(paceinte1)
    medico1 = Medico("Sara", 54621,"si")
    medico1.agregarPaciente(456)
except ValueError as Ve:
    print(f"Error: {Ve}")

       
            
        
        

        
```
